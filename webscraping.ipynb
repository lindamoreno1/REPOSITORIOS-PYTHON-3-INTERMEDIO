```python
import requests
from bs4 import BeautifulSoup
```

```python
pagWeb=requests.get("http://books.toscrape.com/")
soup= BeautifulSoup(pagWeb.text,"html.parser")
print(soup.prettify())
```

```python
titulos=[]
titulo_items=soup.find_all("h3")
for items in titulo_items:
    titulo=items.a
    titulos.append(titulo["title"])
```

```python
titulos
```

```python
precios=[]
precio_items=soup.find_all("p",class_="price_color")
for item in precio_items:
    precio=item.text
    precios.append(precio)
```

```python
precios
```

```python
existencias=[]
existencia_items=soup.find_all("p",class_="instock availability")
for item in existencia_items:
    existencia=item.text
    existencias.append(existencia)
```

```python
existencias
```

```python
pageitems=soup.find_all("p",class_="star-rating ")
print(pageitems)
```

```python
calificacion=[]
estrellas=soup.find_all("p",class_ ="star-rating")
for item in estrellas:
    calificacion.append(item["class"][1])
```
