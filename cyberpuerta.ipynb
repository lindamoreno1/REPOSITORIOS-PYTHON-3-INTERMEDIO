```python
import requests
from bs4 import BeautifulSoup

pagWeb=requests.get("https://www.cyberpuerta.mx/Nuevos-articulos/")
soup= BeautifulSoup(pagWeb.text,"html.parser")
print(soup.prettify())
```

# 

```python
titulos=[]
titulo_items=soup.find_all("div",class_="emproduct_right")
for item in titulo_items:
    titulo=item.a.text.strip()
    titulos.append(titulo)
precios=[]
precio_items=soup.find_all("div",class_="emproduct_right")
for item in precio_items:
    precio=item.label.text.strip()
    precios.append(precio)
```
