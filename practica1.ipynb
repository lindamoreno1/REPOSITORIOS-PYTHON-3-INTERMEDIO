```python
import matplotlib.pyplot as plt
```

```python
xe=dfcopia["Age"].unique()
ye=dfcopia["Age"].value_counts().to_list()
plt.title("Rango de Edad")
plt.xlabel("Edad")
plt.ylabel("Personas")
plt.bar(xe,ye)
plt.show()
```

```python
x=dfcopia["Gender"].unique()
y=dfcopia["Gender"].value_counts().to_list()
plt.title("GENEROS")
plt.xlabel("Genero")
plt.ylabel("Personas")
plt.plot(x,y,".-")
plt.show()
```

```python
xm=dfcopia[(dfcopia["Gender"]=="Male")]["Age"].unique()
ym=dfcopia[(dfcopia["Gender"]=="Male")]["Age"].value_counts().tolist()

xf=dfcopia[(dfcopia["Gender"]=="Female")]["Age"].unique()
yf=dfcopia[(dfcopia["Gender"]=="Female")]["Age"].value_counts().tolist()

xo=dfcopia[(dfcopia["Gender"]=="Other")]["Age"].unique()
yo=dfcopia[(dfcopia["Gender"]=="Other")]["Age"].value_counts().tolist()

#PLOT MUJERES
plt.title("Mujeres")
plt.xlabel("Edad")
plt.ylabel("Cantidad")
plt.subplot(1,3,1)
plt.plot(xf,yf,".-")

#PLOT HOMBRES
plt.title("Hombres")
plt.xlabel("Edad")
plt.ylabel("Cantidad")
plt.subplot(1,3,2)
plt.plot(xm,ym,".-")

#PLOT OTROS
plt.title("Otros")
plt.xlabel("Edad")
plt.ylabel("Cantidad")
plt.subplot(1,3,3)
plt.plot(xo,yo,".-")

plt.suptitle("GENEROS POR EDADES")
plt.figure(figsize=(10,6))
plt.show()
```

```python
width = 0.1
plt.bar(xm-width,ym, width=width, label='Male')
plt.bar(xf,yf, width=width, label='Fenale')
plt.bar(xo-width,yo, width=width, label='Other')

plt.xticks(xe, ye)
plt.legend(loc='best')
plt.figure(figsize=(10,6))
#plt.figure(figsize(10, 6))
plt.show()
```
